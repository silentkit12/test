```python
import requests
import time

LIST_API = "https://korean.visitkorea.or.kr/main/areaListAjax.do"

HEADERS = {
    "User-Agent": "Mozilla/5.0",
    "X-Requested-With": "XMLHttpRequest",
    "Referer": "https://korean.visitkorea.or.kr/main/area_list.do"
}

def fetch_cotid_list(max_page=5):
    cotid_set = set()

    for page in range(1, max_page + 1):
        payload = {
            "page": page,
            "pageSize": 12,
            "type": "Place"
        }

        resp = requests.post(LIST_API, headers=HEADERS, data=payload, timeout=10)
        data = resp.json()

        items = data.get("body", {}).get("result", [])
        if not items:
            break

        for item in items:
            if "cotid" in item:
                cotid_set.add(item["cotid"])

        print(f"ğŸ“„ page {page}: {len(items)}ê°œ")

        time.sleep(0.3)

    return list(cotid_set)
```

```python
DETAIL_URL = "https://korean.visitkorea.or.kr/detail/ms_detail.do?cotid={}"


def crawl_detail(driver, cotid):
    url = DETAIL_URL.format(cotid)
    driver.get(url)
    time.sleep(3)

    # ì¥ì†Œëª…
    try:
        title = driver.find_element(By.ID, "topTitle").text
    except:
        title = ""

    # ì£¼ì†Œ
    try:
        address = driver.find_element(
            By.XPATH,
            "//strong[contains(text(),'ì£¼ì†Œ')]/following-sibling::span"
        ).text
    except:
        address = ""

    # ê°œìš”
    try:
        overview = driver.find_element(
            By.CSS_SELECTOR, ".inr_wrap .inr p"
        ).text
    except:
        overview = ""

    # ì´ë¯¸ì§€
    images = driver.find_elements(By.CSS_SELECTOR, "img")
    img_urls = [
        img.get_attribute("data-src") or img.get_attribute("src")
        for img in images
        if img.get_attribute("src")
    ]

    return {
        "cotid": cotid,
        "title": title,
        "address": address,
        "overview": overview,
        "image_count": len(img_urls)
    }
```

```python
from selenium import webdriver

# 1ï¸âƒ£ ë¦¬ìŠ¤íŠ¸
cotid_list = fetch_cotid_list(max_page=3)
print(f"âœ… ìˆ˜ì§‘ëœ cotid: {len(cotid_list)}ê°œ")

# 2ï¸âƒ£ Selenium ì‹œì‘
driver = webdriver.Chrome(service=service, options=chrome_options)

results = []

for idx, cotid in enumerate(cotid_list[:10], 1):
    try:
        data = crawl_detail(driver, cotid)
        results.append(data)
        print(f"[{idx}] âœ… {data['title']}")
    except Exception as e:
        print(f"[{idx}] âŒ {cotid}: {e}")

driver.quit()
```
